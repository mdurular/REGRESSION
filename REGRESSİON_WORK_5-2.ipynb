```python
prices = pd.read_csv('data_house_prices.csv')
prices.head(1)
```

```python
high=prices.corr().SalePrice.abs().sort_values(ascending=False).index[0:11]
high
```

```python
#Örnek-1
Y = prices['SalePrice']
X = prices[['OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea',
       'TotalBsmtSF', '1stFlrSF', 'FullBath', 'TotRmsAbvGrd', 'YearBuilt',
       'YearRemodAdd']]
lrm = linear_model.LinearRegression()
lrm.fit(X, Y)
```

```python
import statsmodels.api as sm

X = sm.add_constant(X)
results = sm.OLS(Y, X).fit()
results.summary()
```

satış fiyatları ile korelasyonu 0,5'ten büyük değişkenlerle oluşturduğumuz modelimizin r squared, aj. r squared, f değeri ile AIC,BIC değerleri anlamlı sonuç vermektedir.

```python
#örnek-2
Y = prices['SalePrice']
X = prices[['OverallQual', 'GrLivArea', 
       'TotalBsmtSF', '1stFlrSF','YearBuilt',
       'YearRemodAdd']]
lrm = linear_model.LinearRegression()
lrm.fit(X, Y)

X = sm.add_constant(X)
results = sm.OLS(Y, X).fit()
results.summary()
```
